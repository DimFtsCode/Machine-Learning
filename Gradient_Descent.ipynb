**Άσκηση 1.2**

Απο MCB σχέση 132 γνωρίζουμε ότι Frobenius Norm:
$\frac{\partial\|X\|_F}{\partial X} =\frac{Tr(XX^H)}{\partial X}$

$Α^T =$Transposed and complex conjugated matrix (Hermitian) αλλα επειδή Img = 0 έχουμε ότι $X^H = X^T$

$\|A-XB\|_F = Tr((A-XB)(A-XB)^T) =Tr((A-XB)(A^T-X^TB^T)) $

$\|A-XB\|_F =Tr(AA^T) - Tr(AB^TX^T)-Tr(XBA^T)+Tr(XBB^TX^T)$

$\frac{\partial\|A-XB\|_F}{\partial X} =\frac{\partial Tr(AA^T)}{\partial X} - \frac{\partial Tr(AB^TX^T)}{\partial X} - \frac{\partial Tr(XBA^T)}{\partial X} + \frac{\partial Tr(XBB^TX^T)}{\partial X}$



1.  $\frac{\partial Tr(AA^T)}{\partial X} = 0$
2.  $\frac{\partial Tr(AB^TX^T)}{\partial X} =  AB^T$(MCB 100)
3.  $\frac{\partial Tr(XBA^T)}{\partial X} = (BA^T)^T=AB^T$(MCB 102)
4.  $\frac{\partial Tr(XBB^TX^T)}{\partial X} = Χ(ΒΒ^Τ)^Τ + Χ(ΒΒ^Τ)= 2XBB^T$(MCB 117)

Από (1),(2),(3) και (4) έχω ότι:

$$\frac{\partial\|A-XB\|_F}{\partial X} = -2AB^T+2XBB^T $$

Άρα θέλω $-2AB^T+2XBB^T = 0$

Εφόσον Β τετραγωνικός και πλήρης βαθμίδας $ΒΒ^Τ$ αντιστρέψιμος

Άρα $Χ = ΑΒ^Τ(ΒΒ^Τ)^{-1}$ (Close Form Solution)


**Άσκηση 2.1α**

Υπολογίζουμε την παράγωγο της f1(x;w) ως προς w0​ και w1.Έχουμε:
* $g(z) = \frac{1}{1 + e^{-z}}$
* $z(x; w) = w_0 + w_1 x$

Παραγωγίζοντας της g(z) .
* $g'(z) = \frac{d}{dz}\left(\frac{1}{1+e^{-z}}\right) = \frac{e^{-z}}{(1 + e^{-z})^2} = g(z) (1-g(z))$
* $ \frac{\partial z}{\partial w_0} = 1$

Κανόνας **αλυσίδας**

* $ \frac{\partial f_1}{\partial w_0} = \frac{\partial g}{\partial z} \frac {\partial z}{\partial w_0}$

Με βάση τα παραπαπάνω προκύπτει η σχέση 1 καθώς  και η σχέση 2 με την διαφόρα οτι $\frac{\partial z}{\partial w_1} = x$

1. $\frac{\partial f_1}{\partial w_0} = f_1(x; w)(1 - f_1(x; w))$

2. $\frac{\partial f_1}{\partial w_1} = f_1(x; w)(1 - f_1(x; w))x$

Υπολογίζουμε την παράγωγο της L1 ως προς w.

3. $\frac{\partial L_1}{\partial w} = -\frac{1}{n} \sum_{i=1}^n \left( y_i \frac{1}{\ln(2) f_1(x_i; w)} \frac{\partial f_1}{\partial w}(x_i; w) + (1 - y_i) \frac{1}{\ln(2)(1 - f_1(x_i; w))} \left(-\frac{\partial f_1}{\partial w}(x_i; w)\right) \right)$

Αντικαθιστούμε με την σχέση 1 και 2 αντίστοιχα στην σχέση 3.

Από την σχέση (1) και (3) έχουμε:

* $\frac{\partial L_1}{\partial w_0} = -\frac{1}{n \ln(2)} \sum_{i=1}^n \left( y_i (1 - f_1(x_i; w)) - (1 - y_i) f_1(x_i; w) \right) = -\frac{1}{n \ln(2)} \sum_{i=1}^n (y_i - f_1(x_i; w)) $

Από την σχέση (2) και (3) έχουμε:

* $\frac{\partial L_1}{\partial w_1} = -\frac{1}{n \ln(2)} \sum_{i=1}^n (y_i - f_1(x_i; w))x_i$

```python


def sigmoid(z):
    return 1/(1+np.exp(-z))

def compute_gradients(x, y, w0, w1, iterations=150, learning_rate=0.001):
    n = len(y)
    L1_history = []

    for i in range(iterations):
        # Υπολογισμός της πρόβλεψης
        z = w0 + w1 * x
        y_pred = sigmoid(z)

        # Υπολογισμός της L1
        L1 = -np.sum(y * np.log2(y_pred) + (1 - y) * np.log2(1 - y_pred)) / n

        # Αποθήκευση ιστορικού για εκτύπωση
        L1_history.append(L1)

        # Παράγωγοι της L1 ως προς w0 και w1
        dL_dw0 = -np.sum(y - y_pred) / (n * np.log(2))
        dL_dw1 = -np.sum(x * (y - y_pred)) / (n * np.log(2))


        # Ενημέρωση των βαρών
        w0 -= learning_rate * dL_dw0
        w1 -= learning_rate * dL_dw1




        # print(f"Iteration {i + 1}: w0 = {w0:.6f}, w1 = {w1:.6f}, L = {L1:.6f}")

    return w0,w1,L1_history

# Δεδομένα παραδείγματος
x = data[0]
y = data[1]
w0, w1 = 0, 1

# Εκτέλεση της συνάρτησης
w0,w1,L1_hist = compute_gradients(x, y, w0, w1)
```

**Άσκηση 2.1β**

Πρέπει πρώτα να βρούμε τις μερικές παραγώγους της $f_2$ ως προς $w_0,w_1,w_2,w_3$

1.  $\frac{\partial f_2}{\partial w_0} = 1$
2.  $\frac{\partial f_2}{\partial w_1} = \sigma(w_2+w_3x)$
3.  $\frac{\partial f_2}{\partial w_2} = w_1 \sigma'(w_2+w_3x) \frac{\partial (w_2+w_3x)}{\partial w_2} = w_1 \sigma'(w_2+w_3x) $
4.  $\frac{\partial f_2}{\partial w_3} = w_1 \sigma'(w_2+w_3x) \frac{\partial (w_2+w_3x)}{\partial w_3} = w_1 x \sigma'(w_2+w_3x)$

Kαθώς και την παράγωγο της συνάρτησης κόστους ως προς το $w_j$

5.  $\frac{\partial L_2}{\partial w_j} = -\frac{2}{n} \sum_{i=1}^n \left(y_i - f_1(x_i;w) \frac{\partial f_1}{\partial w_j}(x_i;w)   \right) $

Αντικαθιστώντας στη (5) την (1) προκύπτει:
$$\frac{\partial L_2}{\partial w_0} = -\frac{2}{n} \sum_{i=1}^n \left(y_i - f_1(x_i;w)\right)$$

Αντικαθιστώντας στην (5) την (2) προκύπτει:
$$\frac{\partial L_2}{\partial w_1} = -\frac{2}{n} \sum_{i=1}^n \left((y_i - f_1(x_i;w))\sigma(w_2+w_3x)\right)$$

Αντικαθιστώντας στην (5) την (3) προκύπτει:
$$\frac{\partial L_2}{\partial w_2} = -\frac{2}{n} \sum_{i=1}^n \left((y_i - f_1(x_i;w))w_1\sigma'(w_2+w_3x)\right)$$

Αντικαθιστώντας στην (5) την (4) προκύπτει:
$$\frac{\partial L_2}{\partial w_3} = -\frac{2}{n} \sum_{i=1}^n \left((y_i - f_1(x_i;w))w_1x\sigma'(w_2+w_3x)\right)$$

```python


def sigmoid(x):
    return np.where(x > 0, x, 0)


def Tsigmoid(x):
    return np.where(x > 0, 1, 0)

def compute_gradients2(X, Y, w0, w1, w2, w3, iterations=150, learning_rate=0.001):
    n = len(Y)
    L2_history = []

    for i in range(iterations):
        # Υπολογισμός της πρόβλεψης
        Υ_pred = w0 + w1 * sigmoid(w2 + w3 * X)

        # Υπολογισμός της L2
        L2 =  (1/n) * np.sum((Y - Υ_pred)**2)

        # Αποθήκευση ιστορικού για εκτύπωση
        L2_history.append(L2)

        # Παράγωγοι της L2 ως προς w0,w1,w2 και w3
        dL_dw0 = -(2/n) * np.sum(Y - Υ_pred)
        dL_dw1 = -(2/n) * np.sum((Y - Υ_pred) * sigmoid(w2+(w3*X)))
        dL_dw2 = -(2/n) * np.sum((Y - Υ_pred) * w1 * Tsigmoid(w2+(w3*X)))
        dL_dw3 = -(2/n) * np.sum((Y - Υ_pred) * w1 * X * Tsigmoid(w2+(w3*X)))

        # Ενημέρωση των βαρών
        w0 -= learning_rate * dL_dw0
        w1 -= learning_rate * dL_dw1
        w2 -= learning_rate * dL_dw2
        w3 -= learning_rate * dL_dw3




        # print(f"Iteration {i + 1}: w0 = {w0:.6f}, w1 = {w1:.6f}, w2 = {w2:.6f},w3 = {w3:.6f},L = {L2:.6f}")

    return w0,w1,w2,w3,L2_history

# Δεδομένα παραδείγματος
X = data2[0]
Y = data2[1]
w0, w1, w2, w3 = 0.5, 1, 2.2, 3.4
L2_hist = []
# Εκτέλεση της συνάρτησης
w0,w1,w2,w3,L2_hist = compute_gradients2(X, Y, w0, w1, w2, w3)
```

```python


def sigmoid(z):
    return 1/(1+np.exp(-z))

def compute_gradients(x, y, w0, w1, iterations=150, learning_rate=0.001):
    n = len(y)
    L1_history = []

    for i in range(iterations):
        # Υπολογισμός της πρόβλεψης
        z = w0 + w1 * x
        y_pred = sigmoid(z)

        # Υπολογισμός της L1
        L1 = -np.sum(y * np.log2(y_pred) + (1 - y) * np.log2(1 - y_pred)) / n

        # Αποθήκευση ιστορικού για εκτύπωση
        L1_history.append(L1)

        # Παράγωγοι της L1 ως προς w0 και w1
        dL_dw0 = -np.sum(y - y_pred) / (n * np.log(2))
        dL_dw1 = -np.sum(x * (y - y_pred)) / (n * np.log(2))


        # Ενημέρωση των βαρών
        w0 -= learning_rate * dL_dw0
        w1 -= learning_rate * dL_dw1


        # Μείωση του learning rate
        learning_rate *= 0.000001



        print(f"Iteration {i + 1}: w0 = {w0:.6f}, w1 = {w1:.6f}, L = {L1:.6f}")

    return w0,w1,L1_history

# Δεδομένα παραδείγματος
x = data[0]
y = data[1]
w0, w1 = 0, 1

# Εκτέλεση της συνάρτησης
w0,w1,L1_hist = compute_gradients(x, y, w0, w1)
```

```python


def sigmoid(x):
    return np.where(x > 0, x, 0)


def Tsigmoid(x):
    return np.where(x > 0, 1, 0)

def compute_gradients2(X, Y, w0, w1, w2, w3, iterations=150, learning_rate=0.001):
    n = len(Y)
    L2_history = []

    for i in range(iterations):
        # Υπολογισμός της πρόβλεψης
        Υ_pred = w0 + w1 * sigmoid(w2 + w3 * X)

        # Υπολογισμός της L2
        L2 =  (1/n) * np.sum((Y - Υ_pred)**2)

        # Αποθήκευση ιστορικού για εκτύπωση
        L2_history.append(L2)

        # Παράγωγοι της L2 ως προς w0,w1,w2 και w3
        dL_dw0 = -(2/n) * np.sum(Y - Υ_pred)
        dL_dw1 = -(2/n) * np.sum((Y - Υ_pred) * sigmoid(w2+(w3*X)))
        dL_dw2 = -(2/n) * np.sum((Y - Υ_pred) * w1 * Tsigmoid(w2+(w3*X)))
        dL_dw3 = -(2/n) * np.sum((Y - Υ_pred) * w1 * X * Tsigmoid(w2+(w3*X)))

        # Ενημέρωση των βαρών
        w0 -= learning_rate * dL_dw0
        w1 -= learning_rate * dL_dw1
        w2 -= learning_rate * dL_dw2
        w3 -= learning_rate * dL_dw3


        learning_rate *= 0.000001

        print(f"Iteration {i + 1}: w0 = {w0:.6f}, w1 = {w1:.6f}, w2 = {w2:.6f},w3 = {w3:.6f},L = {L2:.6f}")

    return w0,w1,w2,w3,L2_history

# Δεδομένα παραδείγματος
X = data2[0]
Y = data2[1]
w0, w1, w2, w3 = 0.5, 1, 2.2, 3.4
L2_hist = []
# Εκτέλεση της συνάρτησης
w0,w1,w2,w3,L2_hist = compute_gradients2(X, Y, w0, w1, w2, w3)
```

```python
def apply_pca(X, num_components):
    # Κανονικοποίηση των δεδομένων (αφαίρεση της μέσης τιμής και διαίρεση με την τυπική απόκλιση)
    X_mean = np.mean(X, axis=1, keepdims=True)
    X_std = np.std(X, axis=1, keepdims=True)
    X_normalized = (X - X_mean) / X_std
    # print(X_mean)
    
    # Υπολογισμός πίνακα συνδιακύμανσης
    covariance_matrix = np.cov(X_normalized)
    #print(covariance_matrix)
    # Υπολογισμός ιδιοτιμών και ιδιοδιανυσμάτων
    eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)
    # print(eigenvalues,eigenvectors.shape)
    # Ταξινόμηση των ιδιοτιμών 
    sorted_index = np.argsort(eigenvalues)[::-1]
    #print(sorted_index)
    # Ταξινόμηση των ιδιοδανυσμάτων ανά στήλη
    sorted_eigenvectors = eigenvectors[:, sorted_index]
    #print(sorted_eigenvectors)
    # Επιλογή των d κορυφαίων ιδιοδιανυσμάτων
    eigenvector_subset = sorted_eigenvectors[:, :num_components]
    # print(eigenvector_subset.shape)
    # Μετατροπή των δεδομένων
    X_reduced = np.dot(eigenvector_subset.T, X_normalized)

    return X_reduced, eigenvector_subset, X_mean, X_std
```

```python
def reconstruct_image(X_reduced, eigenvector_subset, X_mean, X_std, image_index):
    # Reconstruct της εικόνας από τις μειωμένες διαστάσεις
    X_reconstructed = np.dot(eigenvector_subset, X_reduced[:, image_index].reshape(-1, 1))
    
    # Reconstruct της αρχικής κλίμακας
    X_reconstructed = X_reconstructed * X_std + X_mean
    
    return X_reconstructed
```

```python
sets = ['Set_1', 'Set_2', 'Set_3', 'Set_4', 'Set_5']
datasets = {}
for set_number in sets:
    datasets[set_number] = loadImages(folder_path, set_number)

d_values = range(2, 201, 20)  # Από 2 έως 200 με βήμα 20
errors = {set_number: [] for set_number in sets}

for d in d_values:
    for set_number in sets:
        X, labels = datasets[set_number]
        
        # Εφαρμογή PCA
        X_reduced, eigenvector_subset, X_mean, X_std = apply_pca(X, d)
        
        # Ανακατασκευή των εικόνων και υπολογισμός του σφάλματος
        total_error = 0
        for i in range(X.shape[1]):
            X_reconstructed = reconstruct_image(X_reduced, eigenvector_subset, X_mean, X_std, i)
            error = np.linalg.norm(X[:, i].reshape(-1, 1) - X_reconstructed)
            total_error += error
        
        mean_error = total_error / X.shape[1]
        errors[set_number].append(mean_error)

# Βήμα 3: Απεικόνιση των αποτελεσμάτων
plt.figure(figsize=(12, 8))
for set_number in sets:
    plt.plot(d_values, errors[set_number], marker='o', label=set_number)

plt.title('Mean Reconstruction Error vs Number of Principal Components')
plt.xlabel('Number of Principal Components (d)')
plt.ylabel('Mean Reconstruction Error')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def apply_svd(X):
    X_mean = np.mean(X, axis=1, keepdims=True)
    X_std = np.std(X, axis=1, keepdims=True)
    X_centered = (X - X_mean) / X_std
    U, S, Vt = np.linalg.svd(X_centered, full_matrices=False)
    return U, S, Vt, X_mean, X_std
```

```python
set_number = 'Set_1'
X, y = loadImages(folder_path, set_number)

U, S, Vt, X_mean, X_std = apply_svd(X)

top_singular_vectors = U[:, :9]

fig, axes = plt.subplots(3, 3, figsize=(9, 9))
for i in range(9):
    singular_vector_image = top_singular_vectors[:, i].reshape(50, 50)
    ax = axes[i // 3, i % 3]
    ax.imshow(singular_vector_image, cmap='gray')
    ax.set_title(f'Singular Vector {i+1}')
    ax.axis('off')
    
plt.tight_layout()
plt.show()
```
